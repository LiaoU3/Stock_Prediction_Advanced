```
import requests
from bs4 import BeautifulSoup
```

```
link = 'https://histock.tw/stock/rank.aspx?m=0&d=0&p=all'
```

```
r = requests.get(link)
r
```

```
soup = BeautifulSoup(r.text, 'html.parser')
stock_num_list = []
results = soup.find_all('tr')
# print(results)
for result in results:
    # print(result.select_one("td").getText())
    tmp = result.find('td')
    # print(tmp)
    if tmp != None:     
        stock_num_list.append(tmp.getText())
```
