```
import gspread
import numpy as np
from google.colab import auth
from google.colab import files
from google.colab import drive
from oauth2client.client import GoogleCredentials
import pickle
auth.authenticate_user()
credentials = GoogleCredentials.get_application_default()
gc = gspread.authorize(credentials)
```

```
sh_stock_list = gc.open('Stock_List')
worksheet_list = sh_stock_list.worksheets()
worksheet = worksheet_list[-1]
print(worksheet)
list_of_lists = worksheet.get_all_values()
stock_list = []
for list_ in list_of_lists:
    stock_list.append(list_[0])
print(stock_list)
print(len(stock_list))
```

```
sh_all_date = gc.open('Date')
worksheet_all_date = sh_all_date.get_worksheet(0)
print(worksheet_all_date)
```

```
list_of_lists = worksheet_all_date.get_all_values()
# print(list_of_lists)
date_list = []
for list_ in list_of_lists:
    date_list.append(list_[0])
date_list = date_list[1:]
print(date_list)
print(len(date_list))
```

```
# all_stock_history_price = []
# with open('all_stock_history_price.pickle', 'rb') as f:
#     all_stock_history_price = pickle.load(f)

for i in range(890, len(stock_list)):    
    print('No.'+str(i)+'\tcurrent stock : '+ stock_list[i] +'')
    sh = gc.open(stock_list[i])
    worksheet = sh.get_worksheet(0)
    price_list = worksheet.get_all_values()[1:]    
    one_stock_hitory_price = []
    row = 0
    not_started_yet = True
    if price_list[row][0][:-11] == date_list[0]:
        not_started_yet = False
    for i in range(len(date_list)):
        if row == len(price_list):
            one_stock_hitory_price.append([float(i) for i in price_list[row-1][1:]] )
        else:
            if not_started_yet:
                one_stock_hitory_price.append([0, 0, 0, 0, 0])
                if price_list[row][0][:-11] == date_list[i+1]:
                    not_started_yet = False
            else:
                    
                if price_list[row][0][:-11] == date_list[i]:
                    one_stock_hitory_price.append([float(i) for i in price_list[row][1:]] )
                    row += 1
                else:
                    one_stock_hitory_price.append([float(i) for i in price_list[row-1][1:]])
    all_stock_history_price.append(one_stock_hitory_price)
    with open('all_stock_history_price.pickle', 'wb') as f:
        pickle.dump(all_stock_history_price, f)
```
