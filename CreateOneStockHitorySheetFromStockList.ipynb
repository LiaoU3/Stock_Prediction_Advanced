```
import gspread
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

```
auth.authenticate_user()
credentials = GoogleCredentials.get_application_default()
gc = gspread.authorize(credentials)
```

```
# You have to uncomment this line before when first using it
# sh = gc.create('Stock_Price')
sh = gc.open('Stock_Price')
```

```
sh_stock_list = gc.open('Stock_List')
print(sh_stock_list)
```

```
worksheet_stock_list = sh_stock_list.get_worksheet(0)
list_of_lists = worksheet_stock_list.get_all_values()
print(list_of_lists)
```

```
stock_list = []
for list_ in list_of_lists:
    stock_list.append(list_[0])
print(stock_list)
```

```
# index = stock_list.index('5701')
# print('The index of 5701:', index)
```

```
for i in range(0, len(stock_list)):
    print('Current stock('+ stock_list[i] +') No. ' + str(i))
    sh = gc.create(stock_list[i])
    sh = gc.open(stock_list[i])
    # worksheet = sh.add_worksheet(title=stock, rows='5000', cols='10')
    worksheet = sh.get_worksheet(0)
    worksheet.update_acell('A1', '=GOOGLEFINANCE("TPE:' + stock_list[i] + '","ALL","2010/1/1",TODAY(),"DAILY")')
    print('done')
```
